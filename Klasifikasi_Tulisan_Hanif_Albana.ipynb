```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np

# ====== LOAD DATASET MNIST ======
print("üì• Memuat dataset MNIST...")
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Shape x_train:", x_train.shape)
print("Shape y_train:", y_train.shape)
print("Shape x_test:", x_test.shape)
print("Shape y_test:", y_test.shape)

# ====== Persiapan Data ======
print("üîÑ Persiapan data untuk CNN...")
x_train_cnn = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test_cnn = x_test.reshape(-1, 28, 28, 1) / 255.0

print("Shape x_train_cnn:", x_train_cnn.shape)
print("Shape x_test_cnn:", x_test_cnn.shape)

# ====== Pembuatan Model CNN ======
print("üõ† Membuat model CNN...")
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# ====== Kompilasi Model ======
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# ====== Ringkasan Model ======
print("üìä Ringkasan Model CNN:")
model.summary()

# ====== Pelatihan Model ======
print("üöÄ Melatih model CNN...")
history = model.fit(x_train_cnn, y_train,
                    epochs=5,
                    validation_split=0.1,
                    verbose=1,
                    batch_size=128)

# ====== Simpan Model ======
print("üíæ Menyimpan model...")
model.save("model_cnn.h5")
print("‚úÖ Model tersimpan sebagai 'model_cnn.h5'")

# Simpan model format JSON + weights
print("üíæ Menyimpan model ke format JSON + Weights...")
model_json = model.to_json()
with open("model_cnn.json", "w") as json_file:
    json_file.write(model_json)

# ‚úÖ perbaikan nama file (.weights.h5)
model.save_weights("model_cnn.weights.h5")
print("‚úÖ Model tersimpan sebagai 'model_cnn.json' + 'model_cnn.weights.h5'")

# ====== Evaluasi Model ======
print("üìà Evaluasi model...")
test_loss, test_acc = model.evaluate(x_test_cnn, y_test, verbose=0)
print("Test accuracy:", test_acc)

# ====== Verifikasi Model ======
print("üîç Verifikasi model yang disimpan...")
try:
    loaded_model = models.load_model("model_cnn.h5")
    loaded_model.summary()

    sample_idx = np.random.randint(0, len(x_test))
    sample_image = x_test_cnn[sample_idx].reshape(1, 28, 28, 1)
    actual_label = y_test[sample_idx]

    prediction = loaded_model.predict(sample_image, verbose=0)
    predicted_digit = np.argmax(prediction)

    print(f"‚úÖ Hasil prediksi: {predicted_digit} (Actual: {actual_label})")

    plt.imshow(sample_image.reshape(28, 28), cmap="gray")
    plt.title(f"Actual: {actual_label}, Predicted: {predicted_digit}")
    plt.show()

except Exception as e:
    print("‚ùå Error dalam verifikasi model:", e)

# ====== Visualisasi Hasil Pelatihan ======
print("üìä Membuat grafik hasil training...")
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc="lower right", fontsize=10)
plt.title("Training vs Validation Accuracy", fontsize=12)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc="upper right", fontsize=10)
plt.title("Training vs Validation Loss", fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("training_history.png")
plt.show()

# ====== Verifikasi Weights ======
print("\nüîÑ Verifikasi model dengan load weights...")
try:
    new_model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    new_model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

    # ‚úÖ perbaikan nama file
    new_model.load_weights("model_cnn.weights.h5")
    print("‚úÖ Weights berhasil dimuat kembali")

    sample_idx = np.random.randint(0, len(x_test))
    sample_image = x_test_cnn[sample_idx].reshape(1, 28, 28, 1)
    actual_label = y_test[sample_idx]

    prediction = new_model.predict(sample_image, verbose=0)
    predicted_digit = np.argmax(prediction)
    confidence = np.max(prediction)

    print(f"‚úÖ Prediksi dengan loaded weights: {predicted_digit} "
          f"(Actual: {actual_label}, Confidence: {confidence:.3f})")

except Exception as e:
    print("‚ùå Error dalam verifikasi weights:", e)
```

```python
# ===== TEST DENGAN GAMBAR SENDIRI =====
from tensorflow.keras import models
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import os

# Load model CNN
model = models.load_model('model_cnn.h5')
print("‚úÖ Model CNN loaded!")

# (Opsional) compile ulang untuk hilangkan warning metrics
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

def preprocess_custom_image(image_path):
    """Preprocess gambar tulisan tangan sendiri"""
    img = Image.open(image_path).convert('L')   # grayscale
    img = ImageOps.invert(img)                 # üîÑ invert warna (putih<->hitam)
    img = ImageOps.autocontrast(img)           # perbaiki kontras
    img = img.resize((28, 28))                 # resize ke 28x28
    arr = np.array(img).astype('float32') / 255.0
    arr = arr.reshape(1, 28, 28, 1)            # reshape ke format CNN
    return arr, img

# Test dengan gambar Anda
image_files = ['1.jpg', '4.jpg', '7.jpg']  # bisa tambah lebih banyak file di sini

for file in image_files:
    if os.path.exists(file):
        arr, img = preprocess_custom_image(file)
        prediction = model.predict(arr)
        predicted_digit = np.argmax(prediction)
        confidence = np.max(prediction)

        print(f"üñº File: {file}")
        print(f"‚û°Ô∏è Prediksi: {predicted_digit}, Confidence: {confidence:.3f}")

        # Tampilkan gambar + hasil prediksi
        plt.imshow(img, cmap='gray')
        plt.title(f"Prediksi: {predicted_digit} (Conf: {confidence:.2f})")
        plt.axis("off")
        plt.show()
    else:
        print(f"‚ùå File {file} tidak ditemukan")
```
